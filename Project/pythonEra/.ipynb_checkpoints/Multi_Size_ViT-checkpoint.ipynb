# Tiny ImageNet ViT (PyTorch + timm)

This notebook implements a custom ~1 Million parameter Vision Transformer for Tiny ImageNet (64x64).

**Architecture:**
* **Library:** `timm` (PyTorch Image Models)
* **Input:** 64x64 images
* **Patch Size:** 8
* **Dim:** 96 (Embedding dimension)
* **Depth:** 8 layers
* **Heads:** 4
* **Params:** ~600k

```python

import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import timm
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import requests
import zipfile
import io
import torchinfo
from torchinfo import summary
from PIL import Image

# Setup Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

## 2. Data Preparation
Tiny ImageNet validation folder needs formatting to work with PyTorch `ImageFolder`.

```python
DATA_DIR = './tiny-imagenet-200'

def download_and_format_tiny_imagenet(root_dir):
    # 1. Download if not exists
    url = "http://cs231n.stanford.edu/tiny-imagenet-200.zip" #alternative to duwe's link, given by CS231n and reccommended by projects
    if not os.path.exists(root_dir):
        print("Downloading Tiny ImageNet...")
        r = requests.get(url)
        z = zipfile.ZipFile(io.BytesIO(r.content))
        z.extractall('./')
        print("Extracted.")
    else:
        print("Dataset already exists.")

    # 2. Format Validation Set (Move images into class folders)
    val_dir = os.path.join(root_dir, 'val')
    val_img_dir = os.path.join(val_dir, 'images')
    val_annotations = os.path.join(val_dir, 'val_annotations.txt')
    
    # Check if already formatted
    if os.path.exists(val_img_dir) and len(os.listdir(val_img_dir)) > 0:
        print("Formatting validation set...")
        with open(val_annotations, 'r') as f:
            for line in f:
                parts = line.split('\t')
                img_name = parts[0]
                class_label = parts[1]
                
                # Create class folder if not exists
                class_folder = os.path.join(val_dir, class_label)
                if not os.path.exists(class_folder):
                    os.makedirs(class_folder)
                
                # Move image
                src = os.path.join(val_img_dir, img_name)
                dst = os.path.join(class_folder, img_name)
                if os.path.exists(src):
                    os.rename(src, dst)
        
        # Cleanup empty images folder
        if not os.listdir(val_img_dir):
             os.rmdir(val_img_dir)
        print("Validation set formatted.")
    else:
        print("Validation set already formatted.")

download_and_format_tiny_imagenet(DATA_DIR)
```

## 3. Dataloaders
We use Patch Size 8, so input is 64x64.

```python
BATCH_SIZE = 128
NUM_WORKERS = 2

# Augmentations
train_transform = transforms.Compose([
    transforms.RandomCrop(64, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(0.2, 0.2, 0.2),
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),
])

# Datasets
train_ds = ImageFolder(os.path.join(DATA_DIR, 'train'), transform=train_transform)
val_ds = ImageFolder(os.path.join(DATA_DIR, 'val'), transform=val_transform)

# Loaders
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

print(f"Training classes: {len(train_ds.classes)}")
print(f"Training samples: {len(train_ds)}")
```

## 4. Define Custom ViT Model (~1M Params)
We use `timm` to create a "Micro" ViT.
* **Embed Dim:** 96
* **Depth:** 8
* **Heads:** 4
* **Patch Size:** 8

```python
# Original ViT model for our project
model = timm.models.vision_transformer.VisionTransformer(
    img_size=64,
    patch_size=8,
    in_chans=3,
    num_classes=200,
    embed_dim=96,    # Kept small to reduce params
    depth=8,         # 8 layers
    num_heads=4,     # 4 heads
    mlp_ratio=2.0,   # Reduced MLP ratio
    qkv_bias=True,
)

model = model.to(device)

# Count parameters
param_count = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Model Parameters: {param_count / 1e6:.2f} Million")
```

## 5. Training Loop
Using AdamW and Cosine Annealing.

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=0.05)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)

num_epochs = 30
train_accs, val_accs = [], []
train_losses, val_losses = [], []

print("Starting training...")
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()
        
        pbar.set_postfix({'loss': running_loss/len(train_loader)})
    
    train_acc = 100 * correct / total
    train_loss = running_loss / len(train_loader)
    
    # Validation
    model.eval()
    val_running_loss = 0.0
    val_correct = 0
    val_total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            val_running_loss += loss.item()
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()
            
    val_acc = 100 * val_correct / val_total
    val_loss = val_running_loss / len(val_loader)
    
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    
    scheduler.step()
    
    print(f"Ep {epoch+1}: Train Acc: {train_acc:.2f}% | Val Acc: {val_acc:.2f}% | Val Loss: {val_loss:.4f}")
```

# Training loop (for multiple sized models)

```python
# reusable training loop for multiple models (different sized ones)
def train_model(model, name="Model"):
    print(f"\n{'='*40}")
    print(f"Training {name}")
    print(f"{'='*40}")
    
    # 1. Setup Training Components
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=0.05)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)
    
    # 2. Loop
    num_epochs = 30
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        
        # Training
        for images, labels in train_loader: 
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        # Simple Validation (Top-1 only for speed during training)
        model.eval()
        val_correct, val_total = 0, 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = outputs.max(1)
                val_total += labels.size(0)
                val_correct += predicted.eq(labels).sum().item()
        
        val_acc = 100 * val_correct / val_total
        scheduler.step()
        
        if (epoch + 1) % 5 == 0:
            print(f"Ep {epoch+1}/{num_epochs}: Train Loss: {running_loss/len(train_loader):.4f} | Val Acc: {val_acc:.2f}%")

    # 3. Final Detailed Evaluation (Top 1, 5, 10)
    print(f"\nRunning Final Evaluation for {name}...")
    model.eval()
    correct_1 = 0
    correct_5 = 0
    correct_10 = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            
            # Get Top-10 predictions
            _, pred = outputs.topk(10, 1, True, True)
            pred = pred.t()
            correct = pred.eq(labels.view(1, -1).expand_as(pred))
            
            # Sum up
            correct_1 += correct[:1].reshape(-1).float().sum().item()
            correct_5 += correct[:5].reshape(-1).float().sum().item()
            correct_10 += correct[:10].reshape(-1).float().sum().item()
            total += labels.size(0)
            
    print(f"-"*30)
    print(f"FINAL RESULTS: {name}")
    print(f"Top-1 Acc:  {100 * correct_1 / total:.2f}%")
    print(f"Top-5 Acc:  {100 * correct_5 / total:.2f}%")
    print(f"Top-10 Acc: {100 * correct_10 / total:.2f}%")
    print(f"-"*30)

    return model, val_acc
```

## 6. Visualization (Attention Maps)
Visualizing the CLS token attention.

```python
def visualize_attention(model, img, device):
    model.eval()
    
    # Monkey-patch the last layer to capture attention
    last_block = model.blocks[-1].attn
    original_forward = last_block.forward
    
    saved_attn = {}
    def forwarding_hook(x):
        B, N, C = x.shape
        qkv = last_block.qkv(x).reshape(B, N, 3, last_block.num_heads, C // last_block.num_heads).permute(2, 0, 3, 1, 4)
        q, k, v = qkv.unbind(0)
        attn = (q @ k.transpose(-2, -1)) * last_block.scale
        attn = attn.softmax(dim=-1)
        saved_attn['map'] = attn
        return original_forward(x)
    
    last_block.forward = forwarding_hook
    
    # Run Inference
    img_tensor = val_transform(img).unsqueeze(0).to(device)
    _ = model(img_tensor)
    
    # Restore
    last_block.forward = original_forward
    
    # Process Map
    attn_map = saved_attn['map'].mean(dim=1).squeeze(0)
    cls_attn = attn_map[0, 1:]
    
    grid_size = int(np.sqrt(cls_attn.shape[0]))
    cls_attn_2d = cls_attn.reshape(grid_size, grid_size).detach().cpu().numpy()
    
    cls_attn_resized = torch.nn.functional.interpolate(
        torch.tensor(cls_attn_2d).unsqueeze(0).unsqueeze(0), 
        size=(64, 64), 
        mode='bilinear'
    ).squeeze().numpy()
    
    # Plot
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(img)
    ax[0].set_title("Original Image")
    ax[0].axis('off')
    
    ax[1].imshow(img, alpha=1)
    ax[1].imshow(cls_attn_resized, cmap='jet', alpha=0.5)
    ax[1].set_title("Attention Map")
    ax[1].axis('off')
    plt.show()

# Run visualization
import random
rand_idx = random.randint(0, len(val_ds))
raw_img = ImageFolder(os.path.join(DATA_DIR, 'val'))[rand_idx][0]
visualize_attention(model, raw_img, device)
# 1. Textual Tree View (Standard PyTorch)
# Shows the nested structure of blocks
print("--- Model Hierarchy ---")
print(model)

# 2. Tabular Summary (Keras-Style)
# Shows layer types, output shapes, and parameter counts per layer
print("\n--- Model Summary Table ---")
summary(model, input_size=(1, 3, 64, 64), 
        col_names=["input_size", "output_size", "num_params", "kernel_size", "mult_adds"],
        verbose=1)
```

```python
# Get validation

def calculate_topk_accuracy(model, dataloader, device, topk=(1, 5, 10)):
    model.eval()
    correct_counts = {k: 0 for k in topk}
    total_samples = 0
    
    # Ensure we look for the max k needed (e.g., top 10)
    max_k = max(topk)
    
    print(f"Evaluating Top-{topk} accuracy...")
    
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc="Eval"):
            images, labels = images.to(device), labels.to(device)
            
            # Forward pass
            outputs = model(images)
            
            # Get the top k predictions (indices)
            # output shape: [batch_size, num_classes]
            # pred shape: [batch_size, max_k]
            _, pred = outputs.topk(max_k, 1, True, True)
            pred = pred.t() # Transpose to [max_k, batch_size]
            
            # Check if true label is in prediction
            # labels.view(1, -1) expands to [1, batch_size]
            # expand_as(pred) expands to [max_k, batch_size]
            # correct tensor is boolean matrix of [max_k, batch_size]
            correct = pred.eq(labels.view(1, -1).expand_as(pred))
            
            # Update counters
            batch_size = labels.size(0)
            total_samples += batch_size
            
            for k in topk:
                # Sum the correct predictions in the first k rows
                correct_k = correct[:k].reshape(-1).float().sum().item()
                correct_counts[k] += correct_k

    # Print results
    print("\n" + "="*30)
    print("VALIDATION RESULTS")
    print("="*30)
    for k in topk:
        acc = 100 * correct_counts[k] / total_samples
        print(f"Top-{k} Accuracy: {acc:.2f}%")
    print("="*30)
    
    return correct_counts

# Run evaluation
_ = calculate_topk_accuracy(model, val_loader, device)
```

```python
# Export weights
import struct
import numpy as np

def export_for_cpp(model, dataloader, num_samples=10):
    model.eval()
    
    # --- Part 1: Export Weights ---
    print("Exporting Weights...")
    weight_path = "nano_vit_weights.bin"
    manifest_path = "weights_manifest.txt"
    
    with open(weight_path, 'wb') as f_bin, open(manifest_path, 'w') as f_txt:
        offset = 0
        for name, param in model.named_parameters():
            # Get data as flat float32 array
            data = param.detach().cpu().numpy().astype(np.float32)
            
            # Write binary
            f_bin.write(data.tobytes())
            
            # Write manifest info: Name, Shape, Offset, TotalBytes
            shape_str = ",".join(map(str, data.shape))
            size_bytes = data.nbytes
            f_txt.write(f"{name}|{shape_str}|{offset}|{size_bytes}\n")
            
            offset += size_bytes
            
    print(f"Saved {weight_path} and {manifest_path}")

    # --- Part 2: Export Samples ---
    print(f"Exporting {num_samples} Samples...")
    
    # Get a batch
    images, labels = next(iter(dataloader))
    
    # Run inference to get ground truth
    with torch.no_grad():
        images = images.to(device)
        outputs = model(images)
    
    # Move to CPU
    images = images.cpu()
    outputs = outputs.cpu()
    
    for i in range(num_samples):
        # 1. Prepare Input Image
        # PyTorch is (3, 64, 64) -> C++ wants (64, 64, 3)
        img_tensor = images[i] # [C, H, W]
        img_hwc = img_tensor.permute(1, 2, 0).numpy().astype(np.float32) # [H, W, C]
        
        # 2. Prepare Output Logits
        logits = outputs[i].numpy().astype(np.float32)
        
        # 3. Save Files
        with open(f"sample_{i}_input.bin", "wb") as f:
            f.write(img_hwc.tobytes())
            
        with open(f"sample_{i}_output.bin", "wb") as f:
            f.write(logits.tobytes())
            
        # Optional: Print first few values for sanity check
        if i == 0:
            print(f"Sample 0 Input (First 5 floats): {img_hwc.flatten()[:5]}")
            print(f"Sample 0 Expectation (First 5 logits): {logits[:5]}")

    print("Export Complete.")

# Run the export using your validation loader
export_for_cpp(model, val_loader, num_samples=10)
```

# Outputing weights properly for the CPP implemntation


```python
import os
import numpy as np

def export_weights_to_cpp(model, output_dir="weights"):
    # 1. Create output directory
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created directory: {output_dir}")
    
    model.eval()
    print(f"Exporting weights to '{output_dir}/'...")
    
    count = 0
    # 2. Iterate over all parameters
    for name, param in model.named_parameters():
        # Get the data as a numpy array
        data = param.detach().cpu().numpy()
        
        # Enforce Float32 (Critical for C++ compatibility)
        data = data.astype(np.float32)
        
        # 3. Generate the Filename expected by ML.cpp
        # Rule: Replace dots with underscores. 
        # Example: 'blocks.0.norm1.weight' -> 'blocks_0_norm1_weight.bin'
        filename = name.replace(".", "_") + ".bin"
        
        # Handle specific edge cases if timm naming differs from your expectation
        # (The default timm names align perfectly with the ML.cpp logic we wrote:
        #  blocks.0.mlp.fc1.weight -> blocks_0_mlp_fc1_weight.bin)
        
        filepath = os.path.join(output_dir, filename)
        
        # 4. Write Binary
        with open(filepath, "wb") as f:
            f.write(data.tobytes())
            
        print(f"  Saved: {filename:<40} Shape: {str(data.shape):<15} Size: {data.nbytes/1024:.2f} KB")
        count += 1

    print("-" * 60)
    print(f"Successfully exported {count} weight files to {output_dir}/")

# Run the export
export_weights_to_cpp(model, output_dir="weights")
```

# Output intermediate layer outputs

```python
import torch
import os
import numpy as np

def export_layer_debug_data_final(model, dataloader, output_dir="debug_outputs"):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created {output_dir}/")

    model.eval()
    
    # 1. Get a single sample image
    images, labels = next(iter(dataloader))
    image = images[0].unsqueeze(0).to(device)
    
    # Save Input Image
    img_save = image.cpu().squeeze(0).permute(1, 2, 0).numpy().astype(np.float32)
    path_img = os.path.join(output_dir, "image_0.bin")
    with open(path_img, "wb") as f:
        f.write(img_save.tobytes())
    print(f"Saved Input Image: {path_img} (Shape: {img_save.shape})")

    # 2. Setup Hooks
    activations = {}
    hooks = []

    def get_activation(name):
        def hook(model, input, output):
            activations[name] = output.detach()
        return hook
    
    def get_input_activation(name):
        def hook(model, input, output):
            activations[name] = input[0].detach()
        return hook

    # --- Hook Layers ---
    # Layer 0: Input to first Block (PatchEmbed + CLS + PosEmbed)
    hooks.append(model.blocks[0].register_forward_hook(get_input_activation("layer_0")))
    
    # Layers 1-8: Transformer Blocks
    for i, block in enumerate(model.blocks):
        hooks.append(block.register_forward_hook(get_activation(f"layer_{i+1}")))
        
    # Layer 9: Final Norm
    hooks.append(model.norm.register_forward_hook(get_activation("layer_9")))
    
    # Layer 11: Head Output (Logits)
    # Note: Layer 10 is TokenExtract (handled manually below)
    hooks.append(model.head.register_forward_hook(get_activation("layer_11")))

    # 3. Run Inference
    with torch.no_grad():
        logits = model(image)
        # Calculate Softmax Probabilities for Layer 12
        probs = torch.nn.functional.softmax(logits, dim=1)

    # 4. Save Files
    for h in hooks: h.remove()

    print("\n--- Exporting Layer Outputs ---")
    keys = sorted(activations.keys(), key=lambda x: int(x.split('_')[1]))
    
    for key in keys:
        data = activations[key].cpu().numpy().astype(np.float32)
        
        # Manual Slice for Token Extract (Layer 10)
        if key == "layer_9":
            token_extract = data[:, 0, :]
            p = os.path.join(output_dir, "layer_10_output.bin")
            with open(p, "wb") as f:
                f.write(token_extract.tobytes())
            print(f"Saved: layer_10_output.bin (Manual CLS Extract) Shape: {token_extract.shape}")

        path = os.path.join(output_dir, f"{key}_output.bin")
        with open(path, "wb") as f:
            f.write(data.tobytes())
        print(f"Saved: {key}_output.bin".ljust(30) + f"Shape: {data.shape}")

    # 5. Save Softmax (Layer 12)
    probs_np = probs.cpu().numpy().astype(np.float32)
    p_soft = os.path.join(output_dir, "layer_12_output.bin")
    with open(p_soft, "wb") as f:
        f.write(probs_np.tobytes())
    print(f"Saved: layer_12_output.bin (Softmax Probabilities)  Shape: {probs_np.shape}")

# Run export
export_layer_debug_data_final(model, val_loader)
```

# Export 1000 Images

```python
import os
import shutil
import numpy as np
import torch

def export_bulk_validation_set(dataloader, count=1000, output_dir="data_export"):
    # Define paths matching C++ expectations
    inputs_dir = os.path.join(output_dir, "image_inputs")
    if os.path.exists(inputs_dir): shutil.rmtree(inputs_dir)
    os.makedirs(inputs_dir)
    
    print(f"Exporting {count} images to {inputs_dir}...")
    
    labels_map = []
    exported_count = 0
    
    # Iterate through dataloader
    for batch_images, batch_labels in dataloader:
        if exported_count >= count: break
        
        for i in range(len(batch_images)):
            if exported_count >= count: break
            
            # 1. Prepare Image (CHW -> HWC for C++)
            img_tensor = batch_images[i] # [3, 64, 64]
            img_hwc = img_tensor.permute(1, 2, 0).numpy().astype(np.float32)
            
            # 2. Save Binary
            filename = f"image_{exported_count}.bin"
            path = os.path.join(inputs_dir, filename)
            with open(path, "wb") as f:
                f.write(img_hwc.tobytes())
                
            # 3. Store Label
            labels_map.append(f"{exported_count} {batch_labels[i].item()}")
            
            exported_count += 1
            if exported_count % 100 == 0:
                print(f"  Saved {exported_count}/{count}...")

    # 4. Save Labels Manifest
    labels_path = os.path.join(output_dir, "labels.txt")
    with open(labels_path, "w") as f:
        f.write("\n".join(labels_map))
        
    print(f"Done. Labels saved to {labels_path}")

# Run export (Create data folder if needed)
export_bulk_validation_set(val_loader, count=1000, output_dir="data")
```

# Import 1000 inferences and report accuracy

```python
import os
import numpy as np
import torch

def calculate_cpp_accuracy(data_dir="data"):
    labels_path = os.path.join(data_dir, "labels.txt")
    outputs_dir = os.path.join(data_dir, "inference_outputs")
    
    if not os.path.exists(outputs_dir):
        print(f"Error: Directory {outputs_dir} not found. Did you run the C++ code?")
        return

    print(f"Loading results from {outputs_dir}...")
    
    # 1. Load Ground Truth Labels
    ground_truth = {}
    with open(labels_path, 'r') as f:
        for line in f:
            idx, label = map(int, line.strip().split())
            ground_truth[idx] = label
            
    total = 0
    top1 = 0
    top5 = 0
    top10 = 0
    
    # 2. Iterate through binaries
    # We expect inference_0.bin, inference_1.bin...
    for idx, true_label in ground_truth.items():
        filename = f"inference_{idx}.bin"
        filepath = os.path.join(outputs_dir, filename)
        
        if not os.path.exists(filepath):
            print(f"Warning: Missing {filename}")
            continue
            
        # Load raw float32 logits/probabilities
        with open(filepath, 'rb') as f:
            preds = np.frombuffer(f.read(), dtype=np.float32)
            
        # Sanity check shape (200 classes)
        if preds.shape[0] != 200:
            print(f"Error: {filename} has shape {preds.shape}, expected 200.")
            continue
            
        # Get Top-K Indices (argsorting in descending order)
        # We perform a full sort (-preds) or use argpartition
        top_k_indices = np.argsort(preds)[::-1][:10]
        
        if true_label == top_k_indices[0]:
            top1 += 1
        if true_label in top_k_indices[:5]:
            top5 += 1
        if true_label in top_k_indices[:10]:
            top10 += 1
            
        total += 1
        
    print("-" * 30)
    print(f"Processed: {total} images")
    print(f"Top-1 Accuracy:  {top1/total*100:.2f}%")
    print(f"Top-5 Accuracy:  {top5/total*100:.2f}%")
    print(f"Top-10 Accuracy: {top10/total*100:.2f}%")
    print("-" * 30)

calculate_cpp_accuracy()
```
